```python
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
def get_derivative(x):
    return derivative(f,x,dx=1e-3,n=1)
    #return (f(x+1e-8)-f(x))/1e-8

def get_second_derivative(x):
    return derivative(f,x,dx=1e-3,n=2)


def plot(x,x_start,label):
    if(x>x_start):
        absc=[i for i in range(x_start,2*int(x)-x_start)]
        ordinate=[f(i) for i in range(x_start,2*int(x)-x_start)]
    else:
        absc=[i for i in range(x_start,2*int(x)-x_start,-1)]
        ordinate=[f(i) for i in range(x_start,2*int(x)-x_start,-1)]
        
    plt.plot(absc,ordinate,label='Function')
    plt.plot(x,f(x),'ro',label=label)
    plt.legend(loc='upper right')



#https://en.wikipedia.org/wiki/Newton%27s_method
def find_root(x_start):
    x=x_start
    while True:
        if abs(f(x))<1e-5:
            break
        x-=(f(x)/get_derivative(x))
    plot(x,x_start,'Root')

def find_min_max(x_start):
    x=x_start
    prev_x=0
    while True:
        if abs(prev_x-x)<1e-5:
            break
        prev_x=x
        x-=(get_derivative(x)/get_second_derivative(x))
    plot(x,x_start,'Min-Max')
```

```python
plt.clf()

def f(x):
    return x**3+2*x

print find_root(-1000)
plt.show()
```
